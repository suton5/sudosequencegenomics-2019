```python
def asciitodecimal(string):
    string1=[ord(c) for c in string]
    return string1
```

```python

q3['Q Values'] = q3['Q Values'].apply(asciitodecimal)
```

```python
def dectop(array):
    array1=[10**(-((c-33)/10)) for c in array]
    return array1
```

```python
q3['Q Values'] = q3['Q Values'].apply(dectop)
```

```python
def maxkeeper(array):
    maxq=max(array)
    return maxq
```

```python
q3['Q Values'] = q3['Q Values'].apply(maxkeeper)
```
